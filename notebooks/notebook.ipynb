## Helper functions

```python
#helper function to chunk the sentenses based on a grammar
def parse_sents(ip_sent, grammar):
    cp = nltk.RegexpParser(grammar)
    result = cp.parse(ip_sent)
    return result

#helper function to extract subtrees
def extract_tree(sents,grammar, match):
    ret_list = []
    for sent in sents:
        res = parse_sents(sent, grammar)
        ret_list += [' '.join([word for(word, tag) in subtree]) for subtree in res.subtrees(filter = lambda t: t.node == match)]
    return ret_list
```

##Feature Extraction Functions

    There are 3 regression models one for each category label - Informative, Confusing, Funny.
    Each have a separate set of features; Confusing and Informative have a lot of overlapping features. The following block includes all the feature extraction functions for the different models.

```python
# Lists of words feature extraction functions will use

NEGATION_LIST = ["no", "not", "never", "can't", "won't"]
CERTAINTY_LIST = ["always", "never", "often", "sometimes"]
INTENSIVE_ADVERBS = ["really", "very", "quite", "special", "absolutely", "completely", "cheerfully", "willingly"]
HEDGE_LIST = ["well", "possibly", "maybe"]
with open("../Corpus/positive_words.txt") as f:
    POSITIVE_WORDS = [line.rstrip() for line in f]
with open("../Corpus/negative_words.txt") as g:                
    NEGATIVE_WORDS = [line.rstrip() for line in g]
```

```python


def min_max_normalize(feature):   
    '''Converts a list of numberical values to a list of values between 0 and 1'''
    minVal = min(feature)
    maxVal = max(feature)
    normalized = []
    for e in feature:
        val = float((e - minVal)) / (maxVal - minVal + 1)
        normalized.append(val)
    return feature


def punctuation_count(talk_words, punc):    
    featureVector = []
    for talk in talk_words:
        num_punc = 0
        for word in talk:
            if word == punc:
                num_punc += 1
        #num_punc = sum([1 for word in talk if punc == word])
        featureVector.append(num_punc)
    return featureVector

def num_techterms(para_words_tagged):    
    featureVector = []
    for talk in para_words_tagged:
        sentence_list = [sent for paragraph in talk for sent in paragraph]    # create a list of sentences (remove paragraphs)
        njn_list = extract_tree(sentence_list, 'NJN: {(<N.*>*<J*>*<N.*>+)+}','NJN')
        npn_list = extract_tree(sentence_list, 'NPN: {(<N.*>+<P.*><N.*>+)+}','NPN')
        num = len(njn_list) + len(npn_list)
        featureVector.append(num)
    return featureVector

def num_first_person_pronoun_chunk(para_words_tagged):    
    featureVector = []
    pro_verb = "(<PR.*>|<N.*P>+)+<TO>?<MD>?<TO>?<VB.*>+"
    VB_grammar = "PROVB: {" + pro_verb + "}"
    for talk in para_words_tagged:
        sentence_list = [sent for paragraph in talk for sent in paragraph]    # create a list of sentences (remove paragraphs)
        pvb_list = extract_tree(sentence_list,VB_grammar, "PROVB")
        featureVector.append(len(pvb_list))
    return featureVector    

def num_words(talk_words):    
    featureVector = []
    for talk in talk_words:
        word_list = [word for sents in talk for word in sents]
        num_words = len(word_list)        
        featureVector.append(num_words)
    return featureVector

def avg_word_length(talk_words):    
    featureVector = []
    for talk in talk_words:
        total_letters = 0
        total_words = 0
        for word in talk:
            if word.lower() not in stopwords.words('english') and word.lower() not in string.punctuation:
                total_letters += len(word)
                total_words += 1
        featureVector.append(total_letters / total_words)
    return featureVector

def num_unique_words(talk_words): 
    featureVector = []
    for talk in talk_words:
        featureVector.append(len(set(talk)))
    return featureVector

def num_unique_big_words(talk_words):    
    featureVector = []
    for talk in talk_words:
        unique_big = sum(1 for word in set(talk) if len(word) >= 10)
        featureVector.append(unique_big)
    return featureVector

def num_neg_words(talk_words):    
    featureVector = []
    for talk in talk_words:
        num_neg = len([1 for word in talk if word in NEGATION_LIST])
        featureVector.append(num_neg)
    return featureVector

def num_certain_words(talk_words):    
    featureVector = []
    for talk in talk_words:
        num_certain = len([1 for word in talk if word in CERTAINTY_LIST])
        featureVector.append(num_certain)
    return featureVector

def num_hedge_words(talk_words):    
    featureVector = []
    for talk in talk_words:
        num_hedge = len([1 for word in talk if word in HEDGE_LIST])
        featureVector.append(num_hedge)
    return featureVector

def num_sents(para_words):    
    featureVector = []
    for talk in para_words:
        number_sents = sum([len(paragraph) for paragraph in talk])
        featureVector.append(number_sents)
    return featureVector

def avg_words_sent(para_words):    
    featureVector = []
    for talk in para_words:
        number_sentences = sum([len(paragraph) for paragraph in talk])
        number_words = sum([len(sent) for paragraph in talk for sent in paragraph])
        avg = (number_words / number_sentences)
        featureVector.append(avg)
    return featureVector

def num_paras(para_words):    
    featureVector = []
    for talk in para_words:
        featureVector.append(len(talk))
    return featureVector

def avg_words_para(para_words):    
    featureVector = []
    for talk in para_words:
        number_paragraphs = len(talk)
        number_words = sum([len(sent) for paragraph in talk for sent in paragraph])
        avg = ( number_words / number_paragraphs )
        featureVector.append(avg)
    return featureVector

def count_pos(talk_words_tagged, pos_tag):    
    """Counts number of occurrences of a particular POS in each talk"""
    featureVector = []
    for talk in talk_words_tagged:
        count = 0
        for word , tag in talk:
            if pos_tag in tag:
                count += 1
        featureVector.append(count)
    return featureVector

def num_topics(b_list):
    featureVector = [b for p,b,w,t in b_list]
    return featureVector

def avg_topic_length(b_list):
    featureVector = []
    for p,b,w,t in b_list:
        if b == 0:
            featureVector.append(p)
        else:
            featureVector.append(p/float(b))
    return featureVector

def num_second_person(talk_words_tagged):    
    featureVector = []
    for talk in talk_words_tagged:
        count = 0
        for word , tag in talk:
            if word.lower() == 'you' or word.lower() == 'your':
                count += 1
        featureVector.append(count)
    return featureVector

def positive_negative_words(talk_words, word_list):    
    """count of positive or negative words"""
    featureVector = []
    for talk in talk_words:
        count = sum([1 for word in talk if word in word_list])
        featureVector.append(count)
    return featureVector
```

## Build Feature Vectors

    Here we have the wrapper funtions which call the corresponding feature functions based on the model.

```python
def confusing_matrix():
    header = ["question_count",
               "exclamation_count",
               "semicolon_count",
               #"num_techterms", commented, needs to be removed later
               #"personal_pronoun",
               #"num_words",
               "avg_word_length",
               "num_unique_words",
               "num_unique_big_words",
               "num_neg_words",
               "num_certain_words",
               "num_hedge_words",
               "num_sents",
               "avg_words_sent",
               "num_paras",
               "avg_words_para",
               "count_adjectives",
               "count_verbs",
               "count_conjunction",
               "count_verbs",
               "num_topics",
               "avg_topic_length",
               'num_second_person'
               ]
    feature_list = []
    
    feature_list.append(min_max_normalize(punctuation_count(talk_words, "?")))
    feature_list.append(min_max_normalize(punctuation_count(talk_words, "!")))
    feature_list.append(min_max_normalize(punctuation_count(talk_words, ";")))
    #feature_list.append(min_max_normalize(num_techterms(talk_words_tagged))) #Commented, needs to be removed
    #feature_list.append(min_max_normalize(num_first_person_pronoun_chunk(talk_words_tagged)))
    #feature_list.append(min_max_normalize(num_words(talk_words)))
    feature_list.append(min_max_normalize(avg_word_length(talk_words)))   
    feature_list.append(min_max_normalize(num_unique_words(talk_words)))
    feature_list.append(min_max_normalize(num_unique_big_words(talk_words)))
    feature_list.append(min_max_normalize(num_neg_words(talk_words)))
    feature_list.append(min_max_normalize(num_certain_words(talk_words)))
    feature_list.append(min_max_normalize(num_hedge_words(talk_words)))
    feature_list.append(min_max_normalize(num_sents(para_words)))
    feature_list.append(min_max_normalize(avg_words_sent(para_words)))
    feature_list.append(min_max_normalize(num_paras(para_words)))
    feature_list.append(min_max_normalize(avg_words_para(para_words)))
    feature_list.append(min_max_normalize(count_pos(talk_words_tagged, "JJ")))
    feature_list.append(min_max_normalize(count_pos(talk_words_tagged, "VB")))
    feature_list.append(min_max_normalize(map(lambda x, y: x + y, count_pos(talk_words_tagged, "CC"), 
                                              count_pos(talk_words_tagged, "IN"))))
    feature_list.append(min_max_normalize(count_pos(talk_words_tagged, "VB")))
    feature_list.append(min_max_normalize(num_topics(tagged_topics)))
    feature_list.append((avg_topic_length(tagged_topics)))
    feature_list.append(min_max_normalize(num_second_person(talk_words_tagged)))
    df = pandas.DataFrame(feature_list)
    df.index = header
    df = df.transpose()
    df['Label'] = map(int, confusing_y)
    return df
    
def funny_matrix():
    header = [   "count_interjections",
                 "count_intensive_adverbs",
                 "count_positive_words",
                 "count_negative_words",
                 "count_adjectives",
                 "count_verbs",
                 "count_questions",
                 "count_exclamations"
                ]
    feature_list = []
    
    feature_list.append(min_max_normalize(count_pos(talk_words_tagged, "UH")))
    feature_list.append(min_max_normalize(positive_negative_words(talk_words, INTENSIVE_ADVERBS)))
    feature_list.append(min_max_normalize(positive_negative_words(talk_words, POSITIVE_WORDS)))
    feature_list.append(min_max_normalize(positive_negative_words(talk_words, NEGATIVE_WORDS)))
    feature_list.append(min_max_normalize(count_pos(talk_words_tagged, "JJ")))
    feature_list.append(min_max_normalize(count_pos(talk_words_tagged, "VB")))
    feature_list.append(min_max_normalize(punctuation_count(talk_words, "?")))
    feature_list.append(min_max_normalize(punctuation_count(talk_words, "!")))
    
    df = pandas.DataFrame(feature_list)
    df.index = header
    df = df.transpose()
    df['Label'] = map(int, funny_y)
    return df

def informative_matrix():
    header = ["question_count",
               "exclamation_count",
               "semicolon_count",
               #"num_techterms", commented, needs to be removed later
               #"personal_pronoun",
               #"num_words",
               "avg_word_length",
               "num_unique_words",
               "num_unique_big_words",
               "num_neg_words",
               "num_certain_words",
               "num_hedge_words",
               "num_sents",
               "avg_words_sent",
               "num_paras",
               "avg_words_para",
               "count_adjectives",
               "count_verbs",
               "count_conjunction",
               "count_verbs",
               "num_topics",
               "avg_topic_length",
               'num_second_person'
               ]
    feature_list = []
    
    feature_list.append(min_max_normalize(punctuation_count(talk_words, "?")))
    feature_list.append(min_max_normalize(punctuation_count(talk_words, "!")))
    feature_list.append(min_max_normalize(punctuation_count(talk_words, ";")))
    #feature_list.append(min_max_normalize(num_techterms(talk_words_tagged))) #Commented, needs to be removed
    #feature_list.append(min_max_normalize(num_first_person_pronoun_chunk(talk_words_tagged)))
    #feature_list.append(min_max_normalize(num_words(talk_words)))
    feature_list.append(min_max_normalize(avg_word_length(talk_words)))   
    feature_list.append(min_max_normalize(num_unique_words(talk_words)))
    feature_list.append(min_max_normalize(num_unique_big_words(talk_words)))
    feature_list.append(min_max_normalize(num_neg_words(talk_words)))
    feature_list.append(min_max_normalize(num_certain_words(talk_words)))
    feature_list.append(min_max_normalize(num_hedge_words(talk_words)))
    feature_list.append(min_max_normalize(num_sents(para_words)))
    feature_list.append(min_max_normalize(avg_words_sent(para_words)))
    feature_list.append(min_max_normalize(num_paras(para_words)))
    feature_list.append(min_max_normalize(avg_words_para(para_words)))
    feature_list.append(min_max_normalize(count_pos(talk_words_tagged, "JJ")))
    feature_list.append(min_max_normalize(count_pos(talk_words_tagged, "VB")))
    feature_list.append(min_max_normalize(map(lambda x, y: x + y, count_pos(talk_words_tagged, "CC"), 
                                              count_pos(talk_words_tagged, "IN"))))
    feature_list.append(min_max_normalize(count_pos(talk_words_tagged, "VB")))
    feature_list.append(min_max_normalize(num_topics(tagged_topics)))
    feature_list.append((avg_topic_length(tagged_topics)))
    feature_list.append(min_max_normalize(num_second_person(talk_words_tagged)))
    df = pandas.DataFrame(feature_list)
    df.index = header
    df = df.transpose()
    df['Label'] = map(int, informative_y)
    return df
```

## Split Data

    The data is split into training, development and test sets.

```python
def create_feature_sets(data_frame, split=[.7,.2,.1]):
    trainlen = int(split[0] * len(data_frame))
    devlen = int(split[1] * len(data_frame))
    return data_frame[:trainlen], data_frame[trainlen:trainlen+devlen], data_frame[trainlen:trainlen+devlen:]

df_confusing = confusing_matrix()
df_funny = funny_matrix()
df_informative = informative_matrix()

train_set_con, dev_set_con, test_set_con = create_feature_sets(df_confusing)
train_set_fun, dev_set_fun, test_set_fun = create_feature_sets(df_funny)
train_set_inf, dev_set_inf, test_set_inf = create_feature_sets(df_informative)
```

##Discourse Analysis
    Reference: Structure of Great Talks - Nancy Duarte - http://www.ted.com/talks/nancy_duarte_the_secret_structure_of_great_talks
    The structure of a good talk as per Duarte's analysis follows a pattern of WHAT IS - WHAT COULD BE. The following is an attempt to plot this structure and analyse if there is a pattern that can be seen across different categories of talks.    
     
     
     The following is a talk by Anjelina Jolie at the U.N. This talk is said to be modelled based on this structure.

```python
AJ_talk = ' '.join(open('../Corpus/AJ.txt', 'rb').readlines())
AJ_talk = AJ_talk.replace('\r\n' , '')
sent_tokenized = nltk.sent_tokenize(AJ_talk)
word_tokenized = [nltk.word_tokenize(sent) for sent in sent_tokenized]
tagged = [nltk.pos_tag(word) for word in word_tokenized]
```

    Here, we have picked a set of talks from different categories and plotted the structure. The user ratinf percentages for each label are mentioned alongside the talks.
    'What is' and 'What could be' are differentiated based on the number of past/present tense verbs vs the number of future tense verbs. To get rid of noise, if there is only a minor difference between the two, that shift is ignored. A few additions like 'yet to' , 'responsibility' etc clearly indicate a mention of 'What could be'. So the presence of these words are weighted more.
    
###Plots:
    "What is" is coded at value 5 and "What could be" at 10. The graphs plot sentence boundaries vs the 'What is - What could be' measure. The legend of the plot includes the category percentage for the talk and the number of small shifts that were ignored.

    Initially, we tried the structure analysis on topic boundaries. But that did not capture well the shifts within a topic. So, we moved to a smaller unit - the sentence boundaries.

```python
tagged_sentences = list()
for talk in para_words_tagged:
    tagged_sentences.append([sent for para in talk for sent in para])

talks = [
        (-1, tagged),               # Sample Jolie talk picked from http://www.duarte.com/blog/category/speech-analysis/
        (12, tagged_sentences[12]),     # Informative : 33 vs confusing: 1 vs Funny 3
        (144, tagged_sentences[144]),   # Informative : 9 vs confusing: 4 vs Funny 4
        (151, tagged_sentences[151]),   # Informative : 0 vs confusing: 50 vs Funny 6
        (162, tagged_sentences[162]),   # Informative : 25 vs confusing: 0 vs Funny 3
        (75, tagged_sentences[75]),
        (129, tagged_sentences[129]),    #Inspiring
        ]

trends = [10,5]
def verb_tense(sent):
    loop_pres = list()
    loop_future = list()
    for w,t in sent:
        if 'VB' in t:
            if t == 'VBD' or t == 'VBG' or t == 'VBN' or t == 'VBP' or t == 'VBZ':
                loop_pres.append((w,t))
            else:
                loop_future.append((w,t))
    words = ' '.join([w.lower() for w,t in sent])
    if 'yet to' in words or 'responsibility' in words or 'therefore' in words:
        loop_pres = []
    return len(loop_pres) - len(loop_future)

for index, talk in talks:
    isPresent = None
    small_shift = 0
    trend_list = list()
    for sent in talk:
        present_index = verb_tense(sent)
        if isPresent is None:
            if present_index >= 0:
                isPresent = 1
            else:
                isPresent = 0
        if abs(present_index) < 2:
            small_shift += 1
            pass
        elif isPresent == 1 and present_index < 0:
            isPresent = 0
        elif isPresent == 0 and present_index > 0:
            isPresent = 1
#         if isPresent == 0:
#             print "\t\t\t\t", 
#         print ' '.join([w for w,t in sent])
        trend_list.append(trends[isPresent])
    x = range(len(talk))
    label_text = "What is vs What could be"
    if index != -1:
        label_text = "Confusing: " + confusing_y[index] + "\nFunny: " + funny_y[index] + "\nInformative: " + informative_y[index] + \
                "\nShifts ignored: "  + str(small_shift)
    pylab.xlabel("Sentence Numbers - " + str(len(talk)))
    pylab.ylabel("What is, What could be")
    ax = pylab.subplot(111)
    pylab.plot(x, trend_list, label= label_text)
    ax.legend(bbox_to_anchor=(2, 0.5))
    pylab.show()
```
